```python
import os
import sys
import pandas as pd
import tomllib

# =============================
# IMPORTAR CONEXION
# =============================

sys.path.append(os.path.abspath("../.secrets"))

from db_config import get_connection

print("Importacion OK")


# =============================
# LEER CONFIG
# =============================

with open("../.secrets/secrets.toml", "rb") as f:
    config = tomllib.load(f)

ruta_relativa = config["paths"]["bronze_path"]

# carpeta base del proyecto (sube un nivel desde notebook)
BASE_DIR = os.path.abspath("..")

ruta_archivos = os.path.join(BASE_DIR, ruta_relativa)

print("Ruta relativa:", ruta_relativa)
print("Ruta final:", ruta_archivos)


# =============================
# VERIFICAR RUTA
# =============================

if not os.path.exists(ruta_archivos):
    raise FileNotFoundError(f"No existe la ruta: {ruta_archivos}")

print("Ruta OK")


# =============================
# CONEXION SUPABASE
# =============================

conn = get_connection()
cursor = conn.cursor()

print("Conexion exitosa")


# =============================
# LISTAR CSV
# =============================

archivos = [
    f for f in os.listdir(ruta_archivos)
    if f.endswith(".csv")
]

print("\nArchivos disponibles:")

for i, archivo in enumerate(archivos):
    print(f"{i} - {archivo}")


# =============================
# SELECCIONAR ARCHIVO
# =============================
# ⚠️ en notebook mejor no usar input

indice = 0

print("Indice elegido:", indice)

if indice < 0 or indice >= len(archivos):
    raise ValueError("Indice invalido")


archivo_seleccionado = archivos[indice]

print("Cargando archivo:", archivo_seleccionado)


# =============================
# RUTA COMPLETA
# =============================

ruta_completa = os.path.join(ruta_archivos, archivo_seleccionado)

print("Ruta completa:", ruta_completa)


# =============================
# LEER CSV
# =============================

df = pd.read_csv(ruta_completa)

print("\nPrimeras filas:")
print(df.head())
```

```python
tabla = "bronze.ventas_bronze"

with open(ruta_completa, "r", encoding="utf-8") as f:
    cursor.copy_expert(
        f"""
        COPY {tabla}
        FROM STDIN
        WITH CSV HEADER
        """,
        f
    )

conn.commit()

print("Carga rápida completada")
```
